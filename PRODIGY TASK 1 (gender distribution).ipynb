IMPORTED ALL THE IMPORTANT LIBRARIES NEEDED FOR THE DISTRIBUTION

```python
import pandas as pd
import matplotlib.pyplot as plt


data=pd.read_csv(r"C:\Users\KABEERJR\Downloads\Population of India.csv")
data
```

FILTERED THE ROW TOTAL SINCE ITS IS NOT OF USE IN THE GRAPH

```python
data.info
data.drop(data.index[36], inplace=True)
data.info
```

FOR CREATING A GROUPED BAR GRAPH FOR GENDER DISTRIBUTION ACROSS STATES OF INDIA

```python


state_populations = data.groupby('State/UT').agg({'Population':'sum','Male': 'sum', 'Female': 'sum'})
states = state_populations.index.tolist()


total_population = state_populations['Population'].tolist()
male_population = state_populations['Male'].tolist()
female_population = state_populations['Female'].tolist()


plt.figure(figsize=(10,10))
bar_width = 0.50
index = range(len(states))


plt.bar(index, male_population, bar_width, label='Male Population')
plt.bar([i + bar_width for i in index], female_population, bar_width,label='Female Population')


plt.xlabel('states/UT')
plt.ylabel('population in millions')
plt.title('Gender Distribution in Population Across States')
plt.xticks(index, states,rotation='vertical')
plt.legend()

plt.tight_layout()
plt.show()
```

FOR CREATING A HISTOGRAM FOR POPULATION DISTRIBUTION FOR ALL STATES

```python

population_all_states = data['Population']

# Plotting histogram for population distribution for all states combined
plt.figure(figsize=(10, 6))  # Adjust figure size if needed

plt.hist(population_all_states, bins=20, alpha=0.5, color='blue')

plt.xlabel('Population in millions')
plt.ylabel('Frequency of states/UT')
plt.title('Population Distribution Histogram for All States')
plt.grid(True)
plt.show()

```
